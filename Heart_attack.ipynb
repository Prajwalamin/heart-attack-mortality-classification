# **Classification of 1-year patient mortality following a heart attack**

The provided data set contains data from patients who all suffered heart attacks at some point in the past. Some are still alive and some are not. The data provided contains key clinical information (features) for each patient and the prediction task involves identifying (classifying) which patients are likely to survive for at least one year following the heart attack.
The provided features (clinical variables) to be used as predictors by your classification models include the following:

    1. age-at-heart-attack -- age in years when heart attack occurred
    2. pericardial-effusion -- binary. Pericardial effusion is fluid
			      around the heart.  0=no fluid, 1=fluid
    3. fractional-shortening -- a measure of contracility around the heart
			       lower numbers are increasingly abnormal
    4. epss -- E-point septal separation, another measure of contractility.  
	      Larger numbers are increasingly abnormal.
    5. lvdd -- left ventricular end-diastolic dimension.  This is
	      a measure of the size of the heart at end-diastole.
	      Large hearts tend to be sick hearts.

    6. wall-motion-index -- equals wall-motion-score divided by number of
			   segments seen.  Usually 12-13 segments are seen
			   in an echocardiogram.  
               
The target variable is encoded as a binary outcome of whether a patient survived for 1 year post-heart attack or not. Label '0' indicates that the patient died within one year of a heart attack. Label '1' indicates that the patient survived for at least one year after a heart attack.

## Outliers

We will use the **IQR** method to detect the outliers present in each of the variables. The outliers will be determined as the values which lie below the *second quartile* and after the *third quartile*. Then, we will replace these outliers by the respective column's median value, instead of removing them.

```python
# Excluding 'PericardialEffusion' since it is a binary variable (0 or 1)
heart_outlier = heart.drop(['PericardialEffusion'], axis=1)
pericardial = heart["PericardialEffusion"]
```

```python
# NOTE:- Please run the above code once again before running this cell for the second time as
#        it leads to indifferent results. The reason being, the interquartile range of the data changes
#        everytime this code has been executed. This creates misinterpretation for the function
#        while dealing with outliers.

num_outliers = {}
outlier_index = {}

np.random.seed(123)

def outlier(data):
# identify outliers in each column
    for col in data.columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5*IQR
        upper_bound = Q3 + 1.5*IQR
    
        # Create a mask for outliers
        outlier_mask = (data[col] < lower_bound) | (data[col] > upper_bound)
        
        # no of outliers
        num_outliers[col] = sum(outlier_mask)
        
        # index where the outlier is present
        outlier_index[col] = data.index[outlier_mask]
        
        print(col, "=", len(outlier_index))
        
        median = data[col].median()
        data.loc[outlier_mask, col] = median

#     print(outlier_index)
    
```

```python
print("The number of outliers present in the columns are: \n")
outlier(heart_outlier)
```

# **Exercise 2**

Create a machine learning pipeline using scikit-learn and pre-process the provided data appropriately (3 marks)

```python
# Creating a train and test split from the data.

from sklearn.model_selection import train_test_split


heart_train_X, heart_test_X, heart_train_Y, heart_test_Y = train_test_split(heart_process,heart_targets, test_size=0.2, random_state=42)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator):
    
    def __init__(self, attribute_names):
        self.attribute_names= attribute_names
        
    def fit(self,X, y = None):
        return self
    
    def transform(self, X):
        return X[self.attribute_names].values
```

```python
# Creating a pipeline

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler()),
])


```

```python
# Creating a Pipeline to pre-process the features
from sklearn.pipeline import FeatureUnion
from sklearn.preprocessing import StandardScaler, FunctionTransformer


# Define the columns to be used in the pipeline
num_attr = ['AgeAtHeartAttack','FractionalShortening', 'epss', 'lvdd', 'WallMotionIndex']
cat_attr = ['PericardialEffusion']


# Pipeline for numerical variables
numerical_pipeline = Pipeline(steps=[
    ('selector', DataFrameSelector(num_attr)),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
])

# Pipeline for categorical varibles
categorical_pipeline = Pipeline(steps=[
    ('selector',DataFrameSelector(cat_attr)),
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('scaler', StandardScaler()),
])

# Combine the pipeline
full_pipeline = FeatureUnion(transformer_list=[
    ("numerical_pipeline",numerical_pipeline),
    ("categorical_pipeline",categorical_pipeline),
])

# Fit the pipeline to the data
heart_train_X_transform = full_pipeline.fit_transform(heart_train_X)

# Transform the test data as well
heart_test_X_transform = full_pipeline.fit_transform(heart_test_X)
```

# **Exercise 3**

Train logistic regression classifiers, with and without L1 and L2 regularisation, using the provided data and compare and evaluate their performance. ## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score


# Calling the model
log = LogisticRegression()

# Training the model
log.fit(heart_train_X_transform, heart_train_Y.values.ravel())

# Making predictions
log_predictions = log.predict(heart_test_X_transform)
```

```python
# Confusion Matrix
print(confusion_matrix(heart_test_Y, log_predictions))
```

```python
# EVALUATION!
print(classification_report(heart_test_Y, log_predictions))

# Accuracy of the model
accuracy_log = accuracy_score(heart_test_Y, log_predictions)
print("Accuracy:", round(accuracy_log*100, 2), "%")
```

## Regularization

The accuracy produced by the model is around 74%, but it is also a good idea to consider that there might be some cases of overfitting, which can be reduced by regularization. Hence, we can apply L1 & L2 regularization which aim to solve the problem of overfitting by implementing a **penalty** measure. Therefore, we will train a logistic regression model using *GridSearchCV* to find the optimal regularization weight to be used with the penalty measures **L1** and **L2**. The values for their weights that will be passed are: 10, 1, 0.2, 0.1 and 0.001. 
Then once the model decides the best values out of these, we will train these two models seperately using their respective weights that were identified.


```python
# Grid Search for regularization
from sklearn.model_selection import GridSearchCV

#NOTE: Change the 'penalty' value accordingly to find best regularization weight for L1 or L2.

param_grid_regularization = [{'penalty' : ['l1'],
                              'C' : [10, 1, 0.2, 0.1, 0.001],
                              'solver' : ['liblinear']}
                            ]

log_grid = GridSearchCV(log, param_grid_regularization, cv = 5, verbose=True, n_jobs=-1)

log_grid.fit(heart_train_X_transform, heart_train_Y.values.ravel())
```

```python
# Best regularization parameter value (L1 or L2)
log_grid.best_params_
```

## L1 Regularization

We will use the best paramater value for *C* identified from GridSearchCV.

```python
# Calling the model
log_L1 = LogisticRegression(C=10, penalty = 'l1', solver = 'liblinear')

# Training the model
log_L1.fit(heart_train_X_transform, heart_train_Y.values.ravel())

# Making predictions
log_L1_predictions = log_L1.predict(heart_test_X_transform)
```

```python
# EVALUATION!
print(classification_report(heart_test_Y, log_L1_predictions))

# Accuracy of the model
accuracy_log_L1 = accuracy_score(heart_test_Y, log_L1_predictions)
print("Accuracy:", round(accuracy_log_L1*100, 2), "%")
```

## L2 Regularization
We will use the best paramater value for *C* identified from GridSearchCV (Remember that *penalty* is set to *'l2'* in the variable *param_grid_regularization* ).

```python
# Calling the model
log_L2 = LogisticRegression(C=0.001, penalty = 'l2', solver = 'liblinear')

# Training the model
log_L2.fit(heart_train_X_transform, heart_train_Y.values.ravel())

# Making predictions
log_L2_predictions = log_L2.predict(heart_test_X_transform)
```

```python
# EVALUATION!
print(classification_report(heart_test_Y, log_L2_predictions))

# Accuracy of the model
accuracy_log_L2 = accuracy_score(heart_test_Y, log_L2_predictions)
print("Accuracy:", round(accuracy_log_L2*100, 2), "%")
```

The accuracy achieved by this model is around 66% which is low compared to the other two classifiers. Logistic regression with and without L1 regularization have achieved similar accuracy. However, let us evaluate and compare the performace of all three models using *ROC Curve*.

## ROC Curve

The ROC curve describes how well the model behaves while distinguishing between the classes.

```python
# Return the predicted class probabilites for the three the models

log_probs = log.predict_proba(heart_test_X_transform)
log_L1_probs = log_L1.predict_proba(heart_test_X_transform)
log_L2_probs = log_L2.predict_proba(heart_test_X_transform)

log_probs = log_probs[:, 1]
log_L1_probs = log_L1_probs[:, 1]
log_L2_probs = log_L2_probs[:, 1]
```

```python
# Computing auroc and roc values
from sklearn.metrics import roc_curve, roc_auc_score


log_auc = roc_auc_score(heart_test_Y, log_probs)
log_L1_auc = roc_auc_score(heart_test_Y, log_L1_probs)
log_L2_auc = roc_auc_score(heart_test_Y, log_L2_probs)

# print('Logistic Regression: AUROC = %.3f' % (log_auc))
# print('L1 Logistic Regression: AUROC = %.3f' % (log_L1_auc))
# print('L2 Logistic Regression: AUROC = %.3f' % (log_L2_auc))

log_fpr, log_tpr, _ = roc_curve(heart_test_Y, log_probs)
log_L1_fpr, log_L1_tpr, _ = roc_curve(heart_test_Y, log_L1_probs)
log_L2_fpr, log_L2_tpr, _ = roc_curve(heart_test_Y, log_L2_probs)

#Plotting the curve
plt.plot(log_fpr, log_tpr, linestyle='--', label='Logistic Regression (AUROC = %0.3f)' % log_auc)
plt.plot(log_L1_fpr, log_L1_tpr, marker='.', label='L1 Logistic Regression (AUROC = %0.3f)' % log_L1_auc)
plt.plot(log_L2_fpr, log_L2_tpr, marker='.', label='L2 Logistic Regression (AUROC = %0.3f)' % log_L2_auc)

# Title
plt.title('ROC Plot')
# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# Show legend
plt.legend()
# Show plot
plt.show()
```

Therefore, out of the three classifiers we have defined, the logistic regression model without regularization seems to be performing slightly well when considering its accuracy and auroc value. However, all the three models have close auroc values.
